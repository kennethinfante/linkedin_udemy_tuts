# Geometry operations

#### Buffering geometries

```python
from shapely.geometry import Point, LineString, Polygon
```

```python
point = Point(0, 0)
point
```

```python
point_buffer = point.buffer(0.2)
point_buffer
```

```python
point_buffer = point.buffer(1.0)
point_buffer
```

```python
line = LineString([(0, 0), (0, 1), (1, 1)])
line
```

```python
line_buffer = line.buffer(0.2)
line_buffer
```

```python
line_buffer = line.buffer(1.2)
line_buffer
```

```python
square = Polygon([(0, 0), (0, 1), (1, 1), (1,0)])
square
```

```python
square.buffer(0.2)
```

```python
square.buffer(2.0)
```

```python
square.buffer(10.0)
```

#### Centroid

```python
center_line = line.centroid
center_line
```

```python
print(type(center_line))
```

```python
center_square = square.centroid
center_square
```

```python
print(type(center_square))
```

#### Set operations on geometries

```python
circle_1 = Point(0, 0).buffer(2)
circle_2 = Point(3, 0).buffer(2)
circle_1
```

```python
circle_1.union(circle_2)
```

```python
circle_1.intersection(circle_2)
```

```python
circle_1.difference(circle_2)
```

```python
circle_2.difference(circle_1)
```

#### The within command

```python
point = Point(0, 0)
triangle_1 = Polygon([(-1, -1), (0, 1), (1, -1)])
triangle_2 = Polygon([(3, 4), (5, 14), (10, 10)])
```

```python
triangle_1
```

```python
triangle_2
```

```python
point.within(triangle_1)
```

```python
point.within(triangle_2)
```

```python
point.union(triangle_1).union(triangle_2)
```

#### Measuring distance

```python
point.distance(triangle_1)
```

```python
point.distance(triangle_2)
```
