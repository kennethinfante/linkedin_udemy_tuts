```python
import matplotlib.pyplot as plt
```

```python
f, ax = plt.subplots(1,1,figsize=(6,6))

gdf.plot(ax=ax)
```

```python
f, ax = plt.subplots(1,1,figsize=(6,6))

gdf.plot(ax=ax,
        color = 'crimson',
        edgecolor = 'steelblue',
        linewidth = 2,
        alpha = 0.8)
```

```python
f, ax = plt.subplots(1,1,figsize=(6,6))

gdf.plot(ax=ax,
        column = 'geometry_type',
        edgecolor = 'steelblue',
        linewidth = 2,
        alpha = 0.8)
```

```python
f, ax = plt.subplots(1,1,figsize=(6,6))

gdf.plot(ax=ax,
        column = 'geometry_length',
         cmap = 'Reds',
        edgecolor = 'steelblue',
        linewidth = 2,
        alpha = 0.8)
```
